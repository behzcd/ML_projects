# Customer Churn Prediction model

One of the most common problems in business is to determine the mood of the customer.

If a business owner can predict in advance that a customer will not return, will not use the product or service again, he can act to retain the customer and change his mind.

In this project, we predict whether a customer will stay or not by analyzing the data of online store buyers.

## Data Preparation

```python
print(f"The number of non-existent values: {df.isnull().sum().sum()}")

missing_rows = df[df.isnull().any(axis=1)].shape[0]
print(f"The number of rows that do not exist: {missing_rows}")

print(f"The number of rows that do not exist %: {np.round(missing_rows/len(df)*100)}%")
```

33% is too much. But do we use all the columns?

Let's sort out some categorical (nominal) columns with high correlation.

```python
data = df[['Churn','Tenure','Complain','DaySinceLastOrder','CashbackAmount','MaritalStatus','Gender']]
```

```python
print(f"The number of non-existent values: {data.isnull().sum().sum()}")

missing_rows = data[data.isnull().any(axis=1)].shape[0]
print(f"The number of rows that do not exist: {missing_rows}")

print(f"The number of rows that do not exist %: {np.round(missing_rows/len(df)*100)}%")
```

10% rows value is not available. We will drop these lines.

```python
data = data.dropna()
data.shape
```

```python
data['Churn'].value_counts()/len(data)
```

```python
# We convert text columns to numbers
encoded = pd.get_dummies(data)
encoded.head()
```

```python
X = encoded.drop("Churn", axis=1)
y = encoded['Churn']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
```

## ML

### Logistic Regression

```python
# Training
LR_model = LogisticRegression()
LR_model.fit(X_train, y_train)

# Model evaluation
y_pred = LR_model.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
print("Model accuracy:", metrics.accuracy_score(y_test,y_pred))

## confusion matrix
conf_mat = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()

## ROC curve
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```

### Support Vector Machines

```python
# Training
svm_model = SVC()
svm_model.fit(X_train, y_train)

# Model evaluation
y_pred = svm_model.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
print("Model accuracy:", metrics.accuracy_score(y_test,y_pred))

## confusion matrix
conf_mat = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()

## ROC curve
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```

### Decision Tree

```python
# Training
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)

# Model evaluation
y_pred = tree_model.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
print("Model accuracy:", metrics.accuracy_score(y_test,y_pred))

## confusion matrix
conf_mat = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()

## ROC curve
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```

```python
cols = encoded.drop('Churn', axis=1).columns

plt.figure(figsize=(30,20))
plot_tree(tree_model, feature_names=cols, filled=True)
plt.show()
```

### Random Forest

```python
# Modelni training
RF_model = RandomForestClassifier(n_estimators=9)
RF_model.fit(X_train, y_train)

# Model evaluation
y_pred = RF_model.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
print("Model accuracy:", metrics.accuracy_score(y_test,y_pred))

## confusion matrix
conf_mat = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()

## ROC curve
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```

### XGBoost

```python
# Model training
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)

# Model evaluation
y_pred = xgb_model.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
print("Model accuracy:", metrics.accuracy_score(y_test,y_pred))

## confusion matrix
conf_mat = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()

## ROC curve
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```

If we look at the evaluation of all the algorithms we used, the Decision Tree gave the best result. So as the final model, we get the model made in the Decision Tree.
